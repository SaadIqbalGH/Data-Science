### Compiling Customer Liftime Value 
##### RFM Score Calculations
RECENCY (R): Days since last purchase
FREQUENCY (F): Total number of purchases
MONETARY VALUE (M): Total money this customer spent


```python
# to remove the oulier of negative values from grand total sales column
df= df[df['grand_total']>0]
```

```python
print("Datafile starts with timepoint {}".format(df.created_at.min()))
print("Datafile ends with timepoint {}".format(df.created_at.max()))
```

```python
df["created_at"] = pd.to_datetime(df.created_at, cache=True)

df.created_at.max() - df.created_at.min()
```

```python
import datetime as dt
NOW = dt.datetime(2018,8,28) # taken as current date to compare
```

```python
# Creating a RFM table
rfmTable = df.groupby('Customer ID').agg({'created_at': lambda x: (NOW - x.max()).days, 'increment_id': lambda x: len(x), 'grand_total': lambda x: x.sum()})
rfmTable['created_at'] = rfmTable['created_at'].astype(int)
rfmTable.rename(columns={'created_at': 'Recency', 
                         'increment_id': 'Frequency', 
                         'grand_total': 'Monetary_value'}, inplace=True)
```

```python
'''RECENCY (R): Days since last purchase
   FREQUENCY (F): Total number of purchases
   MONETARY VALUE (M): Total money this customer spent'''

rfmTable[rfmTable['Recency'] > -30].head()
```

```python
df[df['Customer ID']==1.0]
```

```python
quantiles = rfmTable.quantile(q=[0.25,0.5,0.75])
quantiles = quantiles.to_dict()
segmented_rfm = rfmTable
# The lowest recency, highest frequency and monetary amounts are our best customers.
def RScore(x,p,d):
    if x <= d[p][0.25]:
        return 1
    elif x <= d[p][0.50]:
        return 2
    elif x <= d[p][0.75]: 
        return 3
    else:
        return 4
    
def FMScore(x,p,d):
    if x <= d[p][0.25]:
        return 4
    elif x <= d[p][0.50]:
        return 3
    elif x <= d[p][0.75]: 
        return 2
    else:
        return 1
```

```python
segmented_rfm['R_quartile'] = segmented_rfm['Recency'].apply(RScore, args=('Recency',quantiles,))
segmented_rfm['F_quartile'] = segmented_rfm['Frequency'].apply(FMScore, args=('Frequency',quantiles,))
segmented_rfm['M_quartile'] = segmented_rfm['Monetary_value'].apply(FMScore, args=('Monetary_value',quantiles,))
segmented_rfm.head()
```

```python
segmented_rfm['RFMScore'] = segmented_rfm.R_quartile.map(str) + segmented_rfm.F_quartile.map(str) + segmented_rfm.M_quartile.map(str)
segmented_rfm.head()
```
